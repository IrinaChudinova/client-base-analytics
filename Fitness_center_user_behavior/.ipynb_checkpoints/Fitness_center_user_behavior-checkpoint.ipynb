## Анализ поведения клиентов сети фитнес-центров

Необходимо провести анализ работы сети фитнес-центров и подготовить план действий для уменьшения оттока клиентов. В рамках проекта необходимо найти факторы, которые наиболее сильно влияют на отток, спрогнозировать вероятность оттока клиента, сформировать портрет клиента

#### Описание данных:
- *Churn* — факт оттока в текущем месяце;

Данные пользователя за предыдущий до проверки факта оттока месяц:
- *gender* — пол
- *Near_Location* — проживание или работа в районе, где находится фитнес-центр
- *Partner* — сотрудник компании-партнёра клуба (сотрудничество с компаниями, чьи сотрудники могут получать скидки на абонемент — в таком случае фитнес-центр хранит информацию о работодателе клиента)
- *Promo_friends* — факт первоначальной записи в рамках акции «приведи друга» (использовал промо-код от знакомого при оплате первого абонемента)
- *Phone* — наличие контактного телефона
- *Age* — возраст
- *Lifetime* — время с момента первого обращения в фитнес-центр (в месяцах)

Информация на основе журнала посещений, покупок и информация о текущем статусе абонемента клиента
- *Contract_period* — длительность текущего действующего абонемента (месяц, 3 месяца, 6 месяцев, год)
- *Month_to_end_contract* — срок до окончания текущего действующего абонемента (в месяцах)
- *Group_visits* — факт посещения групповых занятий
- *Avg_class_frequency_total* — средняя частота посещений в неделю за все время с начала действия абонемента
- *Avg_class_frequency_current_month* — средняя частота посещений в неделю за предыдущий месяц
- *Avg_additional_charges_total* — суммарная выручка от других услуг фитнес-центра: кафе, спорт-товары, косметический и массажный салон

### Шаг 3. Построим модель прогнозирования оттока пользователей

```python
# исключим avg_class_frequency_total и contract_period из матрицы признаков
df_new = df.drop('avg_class_frequency_total', axis = 1)
df_new = df_new.drop('contract_period', axis = 1)
```

```python
# построим модель бинарной классификации пользователей, где целевой признак — факт оттока пользователя в следующем месяце
# разобьем данные на обучающую и валидационную выборку функцией train_test_split()

# разделим наши данные на признаки (матрица X) и целевую переменную (y)
X = df_new.drop(columns = ['churn'])
y = df_new['churn']

# разделяем модель на обучающую и валидационную выборку
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
# создадим объект класса StandardScaler и применим его к обучающей выборке
scaler = StandardScaler()
X_train_st = scaler.fit_transform(X_train) #обучаем scaler и одновременно трансформируем матрицу для обучающей выборки

# применяем стандартизацию к матрице признаков для тестовой выборки
X_test_st = scaler.transform(X_test)
```

```python
# обучим модель на train-выборке двумя способами:
# логистической регрессией
# случайным лесом
# зададим список моделей
models = [LogisticRegression(), RandomForestClassifier()]

# оценим метрики accuracy, precision и recall для обеих моделей на валидационной выборке
# функция, которая принимает на вход модель и данные и выводит метрики
def make_prediction(m, X_train, y_train, X_test, y_test):
    model = m
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    print('Accuracy: {:.2f} Precision: {:.2f} Recall: {:.2f} F1: {:.2f} '.format(accuracy_score(y_test, y_pred),
                                                                      precision_score(y_test, y_pred),
                                                                      recall_score(y_test, y_pred),
                                                                      f1_score(y_test, y_pred)))
# напишите цикл, который выводит метрики по списку моделей
for i in models:
    print(i)
    make_prediction(m=i, X_train=X_train_st, y_train=y_train, X_test=X_test_st, y_test=y_test)
```

Логистическая регрессия незначительно лучше, чем модель случайного леса. Модель логистической регрессии правильно прогнозирует 90% правильных ответов, а случайного леса - 88%. Метрики точности (precision) и полноты (recall) также лучше у модели логистической регрессии. Так как precision и recall направлены на избежание противоположных рисков, оценим F1-score - метрику, учитывающую баланс между метриками. ### Шаг 4. Сделаем кластеризацию пользователей

```python
# отложим в сторону столбец с оттоком и проведем кластеризацию объектов (пользователей):
# стандартизируем данные
# создаём объект класса scaler (нормализатор)
scaler = StandardScaler()
# обучаем нормализатор и преобразуем набор данных
x_sc = scaler.fit_transform(df_new.drop(columns = ['churn'])) 

# построим матрицу расстояний функцией linkage() на стандартизованной матрице признаков и нарисуем дендрограмму
linked = linkage(x_sc, method = 'ward')

# В переменной linked сохранена таблица «связок» между объектами. Её можно визуализировать как дендрограмму
plt.figure(figsize=(10, 8))  
dendrogram(linked, orientation='top')
plt.title('Hierarchial clustering')
plt.show()
```

```python
# на основании полученного графика предположите, какое количество кластеров можно выделить
# предложенное оптимальное число кластеров 4 — четыре разных цветов на графике
# за число кластеров принять n=5 (в рамках обучения, для сравнения с другими работами)
# обучим модель кластеризации на основании алгоритма K-Means и спрогнозируем кластеры клиентов
# задаём модель k_means с числом кластеров 3
km = KMeans(n_clusters = 5, random_state=0)

# прогнозируем кластеры для наблюдений (алгоритм присваивает им номера от 0 до 2)
labels = km.fit_predict(x_sc)

# сохраняем метки кластера в поле нашего датасета
df_new['cluster_km'] = labels

# выводим статистику по средним значениям наших признаков по кластеру
print(df_new.groupby(['cluster_km']).mean())
```

<div style="border:solid black 1px; padding: 20px">
Обратим внимание, что в разных кластерах разное среднее, что говорит о корректной разбивки на кластеры.

```python
# построим распределения признаков для кластеров
for col in df_new.drop('churn', axis = 1).columns:
    sns.distplot(df_new.query('cluster_km==0')[col])
    sns.distplot(df_new.query('cluster_km==1')[col])
    sns.distplot(df_new.query('cluster_km==2')[col])
    sns.distplot(df_new.query('cluster_km==3')[col])
    sns.distplot(df_new.query('cluster_km==4')[col])
    sns.despine(offset=10)
    plt.show()
```

```python
# для каждого полученного кластера посчитаем долю оттока (методом groupby())
# Какие кластеры склонны к оттоку, а какие — надёжны?
churn_cluster = pd.DataFrame()
churn_cluster['count'] = df_new.groupby('cluster_km')['churn'].count()
churn_cluster['sum'] = df_new.groupby('cluster_km')['churn'].sum()
churn_cluster['churn_rate'] = churn_cluster['sum'] / churn_cluster['count'] * 100
print(churn_cluster)
```

<div style="border:solid black 1px; padding: 20px">
Наибольший процент оттока в 3-ем классе, он составляет 74%, также необходимо обратить внимание на 1 и 4 класс, процент оттока в которых соответственно 10,8% и 40,1%
